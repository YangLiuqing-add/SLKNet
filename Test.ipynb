```python
from matplotlib import pyplot as plt
from keras.models import load_model
import scipy.io
import scipy.io
import numpy as np
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
from keras import backend as K
from Utils import *
```

```python
### load trained model
model = load_model('best_model_forge.h5',custom_objects={'tf': tf})
model.summary()
```

```python
### load test data
mat = scipy.io.loadmat ("./Data/FORGE/Test/FORGE_example1.mat")
dn = mat['dn']
testt = mat['dn']

print('The size of test data:',dn.shape)
```

```python
### denoise the test data
w1=64
w2=64
z1=32  # The z1 and z2 can be set to a small size, e.g., 1,2, and 4
z2=32

n1,n2 = np.shape(testt)
test = yc_patch(testt, w1, w2, z1, z2)
test = np.array(test)

test1 = np.reshape(test,(test.shape[0], w1, w2,1))

predicted = model.predict(test1)
```

```python
### Unpatch the denoised patches

predicted = np.reshape(predicted,(predicted.shape[0],w1*w2))
#print(predicted.shape)
predicted = np.transpose(predicted)
predicted = yc_patch_inv(predicted,n1,n2,w1,w2,z1,z2)

#scipy.io.savemat("./Result/FORGE/de_FORGE_example1.mat", mdict={'slknet': predicted})
```

```python
### Plot the noisy data, denoised data, and removed noise

def cseis():
    from matplotlib.colors import ListedColormap
    import numpy as np
    seis=np.concatenate(
(np.concatenate((0.5*np.ones([1,40]),np.expand_dims(np.linspace(0.5,1,88),axis=1).transpose(),np.expand_dims(np.linspace(1,0,88),axis=1).transpose(),np.zeros([1,40])),axis=1).transpose(),
np.concatenate((0.25*np.ones([1,40]),np.expand_dims(np.linspace(0.25,1,88),axis=1).transpose(),np.expand_dims(np.linspace(1,0,88),axis=1).transpose(),np.zeros([1,40])),axis=1).transpose(),
np.concatenate((np.zeros([1,40]),np.expand_dims(np.linspace(0,1,88),axis=1).transpose(),np.expand_dims(np.linspace(1,0,88),axis=1).transpose(),np.zeros([1,40])),axis=1).transpose()),axis=1)
    return ListedColormap(seis)

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(2,2,1)
plt.imshow(dn,cmap=cseis(),clim=(-25,25),aspect=0.2);ax.set_xticks([]);ax.set_yticks([]);
plt.title('Noisy data');
ax = fig.add_subplot(2,2,2)
plt.imshow(predicted,cmap=cseis(),clim=(-25,25),aspect=0.2);ax.set_xticks([]);ax.set_yticks([]);
plt.title('Denoised data');
ax = fig.add_subplot(2,2,3)
plt.imshow(dn-predicted,cmap=cseis(),clim=(-25,25),aspect=0.2);ax.set_xticks([]);ax.set_yticks([]);
plt.title('Removed noise');
plt.show()
```
